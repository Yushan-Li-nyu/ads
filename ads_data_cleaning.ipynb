```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the shapefile
data = gpd.read_file("./ZIP_CODE_040114.shp")
data.head()
```

```python
data.plot()
```

```python
data.columns
```

```python
# keep brooklyn zipcodes
data['ZIPCODE'] = data['ZIPCODE'].astype(int)
data_bk = data[(data['ZIPCODE'] >= 11201) & (data['ZIPCODE'] <= 11256)]
data_bk
```

```python
data_bk.shape
```

```python
data_bk.PO_NAME.unique()
```

```python
columns_to_drop = ['STATE', 'COUNTY', 'SHAPE_AREA', 'SHAPE_LEN']

# Drop the specified columns
data_BK = data.drop(columns=columns_to_drop)
data_BK.shape
```

```python
# no null counts to remove
null_counts = data_BK.isnull().sum()
print(null_counts)
```
