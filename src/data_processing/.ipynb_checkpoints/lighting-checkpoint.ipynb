```python
## The data has alrealdy been narrowed down to Borough == Brooklyn
Data = pd.read_csv('../../data/raw/light_bk.csv')
```

```python
pd.set_option('display.max_columns', None)
```

```python
Data.head(3)
```

```python
print(f'Initial data shape is {Data.shape}')
light_location = Data.drop_duplicates(subset=['Location'])
print(f'After droping the duplicates, the shape is {light_location.shape}')
light_location = light_location.dropna(subset=['Location'])
print(f'After droping NA values, the shape is {light_location.shape}')
light_location.head(3)
```

```python
light_location.columns
```

```python
lights_bk = light_location[['Borough','Incident Zip','Latitude','Longitude','Location', 'Complaint Type', 'Descriptor']]
```

```python
# filter out only Brooklyn data
lights_bk = lights_bk[~lights_bk['Incident Zip'].isna()]
lights_bk['Incident Zip'] = lights_bk['Incident Zip'].astype(int)
lights_bk = lights_bk[(lights_bk['Incident Zip'] >= 11201) & (lights_bk['Incident Zip'] <= 11256)]
print(f'The cleaned Brooklyn landuse data shape is {lights_bk.shape}.')
lights_bk
```

```python
if len(lights_bk['Complaint Type'].unique()) == 1:
    print('Delete column Complaint Type, because it only has one value:', lights_bk['Complaint Type'].unique()[0])
    lights_bk = lights_bk.drop(['Complaint Type'], axis=1)
```
