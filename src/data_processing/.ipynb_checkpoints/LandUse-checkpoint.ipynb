```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../../data/raw/landuse.csv')
df.columns
```

```python
df_cleaned = df.dropna(subset=['landuse'])
# Narrow down to Brooklyn
df_BK_cleaned = df_cleaned[df_cleaned['borough'] == 'BK']
df_BK_cleaned = df_BK_cleaned[~df_BK_cleaned['zipcode'].isna()]
df_BK_cleaned['zipcode'] = df_BK_cleaned['zipcode'].astype(int)
df_BK_cleaned = df_BK_cleaned[(df_BK_cleaned['zipcode'] >= 11201) & (df_BK_cleaned['zipcode'] <= 11256)]
print(f'The cleaned whole dataset shape is {df_cleaned.shape}, and the cleaned Brooklyn landuse data shape is {df_BK_cleaned.shape}.')
df_BK_cleaned
```

```python
aggregated_data = df_BK_cleaned.groupby('landuse')['borough'].agg('count')
aggregated_data
```

```python
## check all columns whether only one unique values included
uniq_col = {}
for col in df_BK_cleaned.columns:
    if len(df_BK_cleaned[col].unique()) == 1:
        uniq_col[col] = df_BK_cleaned[col].unique()[0]
uniq_col
```

**Check the Data Dictionary**

'spdist3': nan\
If the tax lot is not divided by at least three special purpose districts, the field is blank.

```python
df_BK_cleaned = df_BK_cleaned.drop(['spdist3', 'version', 'notes'], axis=1)
```

```python
# 将经纬度转换为Point几何类型
df_BK_cleaned['geometry'] = df_BK_cleaned.apply(lambda row: Point(row['longitude'], row['latitude']), axis=1)

# 将DataFrame转换为GeoDataFrame
gdf_BK_cleaned = gpd.GeoDataFrame(df_BK_cleaned, geometry='geometry')

fig, ax = plt.subplots()

gdf_BK_cleaned.plot(ax=ax, column='landuse', cmap='viridis', legend=True, markersize=1)

plt.show()
```

```python
# Map the 'landuse' values to new categories
def categorize_landuse(value):
    if value in [1, 2, 3]: ## residential buildings
        return 'Residential'
    elif value in [4, 5, 6]: ## business
        return 'Business'
    elif value in [7, 8, 9, 10, 11]: ## public
        return 'Public'
    else:
        return None

# Apply the mapping function to create a new column
df_BK_cleaned['landuse_category'] = df_BK_cleaned['landuse'].apply(categorize_landuse)

df_BK_cleaned['landuse_category'].value_counts()
```

```python
##The total gross area in square feet, except for condominium measurements which 
## come from the Condo Declaration and are net square footage not gross.
df_BK_cleaned.bldgarea
```

```python
landuse_area = df_BK_cleaned.groupby(['zipcode', 'landuse_category'])['bldgarea'].sum().reset_index(name='landuse_area_sqft')

landuse = landuse_area.pivot_table(index='zipcode', columns='landuse_category', values='landuse_area_sqft')
landuse['total_landuse_area_sqft'] = landuse.sum(axis=1)
for col in landuse.columns[:3]:
    landuse[f'{col}_density'] = landuse[col]/landuse['total_landuse_area_sqft']
```

```python
plt.figure(figsize=(12, 8))
landuse.iloc[:,-3:].plot(kind='bar', stacked=True, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'], figsize=(14, 8))
plt.title('Proportion of Landuse Categories by Zipcode')
plt.xlabel('Zipcode')
plt.ylabel('Percentage')
plt.legend(title='Landuse Category', labels=['Category 1', 'Category 2', 'Category 3'], loc='upper right')
plt.tight_layout()
plt.show()
```

```python
## save the file
landuse.to_csv('../../data/processed_data/landuse.csv')
```
