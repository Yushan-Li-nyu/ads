```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the shapefile
data = gpd.read_file("../../data/raw/ZIP_CODE_040114.shp")
data.head(3)
```

```python
data.plot()
```

```python
data.columns
```

```python
# keep brooklyn zipcodes
data['ZIPCODE'] = data['ZIPCODE'].astype(int)
data_bk = data[(data['ZIPCODE'] >= 11201) & (data['ZIPCODE'] <= 11256)]
data_bk.head(3)
```

```python
data_bk.shape
```

```python
column_chosen = []
for col in list(data_bk.columns[1:]):
    if len(data_bk[col].unique()) == 1:
        column_chosen.append(col)
        print(f'There is only 1 unique value in column {col}, which is {data_bk[col].unique()}')
```

```python
columns_to_drop = set(column_chosen) - set(['PO_NAME','ST_FIPS', 'CTY_FIPS'])


# Drop the specified columns
data_bk = data_bk.drop(columns=columns_to_drop)
print(f'Columns dropped: {columns_to_drop}')
print(f'Columns kept: {data_bk.columns}')
print(f'Current data shape: {data_bk.shape}')
```

```python
# no null counts to remove
null_counts = data_bk.isnull().sum()
print(null_counts)
```

```python
data_bk.head(3)
```

```python
data_bk.to_file('../../data/processed_data/data_bk.shp', index=False)
data_bk.to_csv('../../data/processed_data/data_bk.csv', index=False)
```
