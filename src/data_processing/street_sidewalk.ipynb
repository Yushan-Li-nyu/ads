# Sanity Check

```python
# rename the columns
street_sidewalk = street_sidewalk.rename(columns={
    'OBJECTID_12': 'Zipcode_ID', 'ZIPCODE': 'ZIPCODE', 'PO_NAME': 'Zipcode_PO_NAME', 'POPULATION': 'Zipcode_POPULATION', 
                   'AREA': 'Zipcode_AREA', 'ST_FIPS': 'Zipcode_ST_FIPS','CTY_FIPS': 'Zipcode_CTY_FIPS', 'Area_sqft': 'Zipcode_Area_sqft', 
                   'OBJECTID': 'Sidewalk_ID', 'sidewalks_bk_AddSpatialJoin_2_ZIPCODE': 'Sidewalk_ZIPCODE', 
                   'Shape_Length': 'Sidewalk_Shape_Length', 'Shape_Area': 'Sidewalk_Shape_Area', 'OBJECTID_1': 'Street_ID', 
                   'street_bk_AddSpatialJoin_ZIPCODE': 'Street_ZIPCODE', 'FAIR': 'Street_FAIR', 'GOOD': 'Street_GOOD', 'NR': 'Street_NR',
                   'POOR': 'Street_POOR', 'sum_length_ft': 'Street_sum_length_ft', 'fair_ratio': 'Street_fair_ratio', 
                   'good_ratio': 'Street_good_ratio', 'nr_ratio': 'Street_nr_ratio','poor_ratio': 'Street_poor_ratio'
                  }, errors="raise")
```

```python
# sanity check of whether zipcodes for every file before combining are the same, they should be same!
if sum(
    (street_sidewalk['ZIPCODE'] == street_sidewalk['Sidewalk_ZIPCODE']) & 
    (street_sidewalk['Sidewalk_ZIPCODE'] == street_sidewalk['Street_ZIPCODE'])
) != street_sidewalk.shape[0]:
    print('Go back to check the zipcodes for three files beforing combining them into street_sidewalk')
else:
    street_sidewalk = street_sidewalk.drop(['Sidewalk_ZIPCODE', 'Street_ZIPCODE'], axis=1)
```

```python
# sanity chck of sidewalk and street ID, they are the same, and have 2 values counts of ID 29:
if sum((street_sidewalk['Sidewalk_ID'] == street_sidewalk['Street_ID'])) == street_sidewalk.shape[0]:
    street_sidewalk = street_sidewalk.drop(['Street_ID'], axis=1)
```

```python
import geopandas as gpd
zipcode = gpd.read_file("../../data/processed_data/zipcode_bk.shp"); zipcode.head(3)
zipcode_gdf = gpd.GeoDataFrame(zipcode, geometry=zipcode['geometry'])#, crs="EPSG:2263")
```

```python
zipcode_gdf = zipcode_gdf.drop(21)
```

```python
combined_df = pd.merge(zipcode_gdf, street_sidewalk, on='ZIPCODE', how='right')
```

```python
combined_df
```

```python
import contextily as ctx
import matplotlib.pyplot as plt

categories = ['Street_good_ratio', 'Street_fair_ratio', 'Street_poor_ratio', 'Street_nr_ratio']
combined_df.to_crs('EPSG:4326', inplace=True)
combined_df = combined_df.to_crs(epsg=3857)

fig, axs = plt.subplots(2, 2, figsize=(20, 16))
axs = axs.flatten()

for i, category in enumerate(categories):
    ax = axs[i]
    geo_df = gpd.GeoDataFrame(combined_df, geometry='geometry')
    vmin, vmax = geo_df[category].min(), geo_df[category].max()
    geo_df.plot(column=category, cmap='Blues', linewidth=0.6, ax=ax, edgecolor = '0.8',
                legend=True, vmin=vmin, vmax=vmax)
    if category.split("_")[1] == 'nr':
        ax.set_title(f'Sum of Not-Rated Street Length to Total Street Length by Zipcode')
    else:
        ax.set_title(f'Sum of {category.split("_")[1].capitalize()} Street Length to Total Street Length by Zipcode')
    # ctx.add_basemap(ax, crs=geo_df.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik)

plt.suptitle('Sum of Street Length to Total Street Length by Zipcode by Rate Types', fontsize=30)
#plt.tight_layout()
plt.savefig('../../results/street_rate_ratio.png')
plt.show()
```

```python
combined_df['Sidewalk_Density'] = combined_df['Sidewalk_Shape_Area'] / combined_df['AREA']
combined_df = combined_df.to_crs(epsg=3857)

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
combined_df.plot(column='Sidewalk_Density', cmap='viridis', linewidth=0.6, ax=ax, edgecolor='0.8', legend=True)
ctx.add_basemap(ax, crs=combined_df.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik, alpha = 0.5)

ax.set_title('Sidewalk Area Density Heatmap')
plt.savefig('../../results/sidewalk_area_density.png')
plt.show()
```
